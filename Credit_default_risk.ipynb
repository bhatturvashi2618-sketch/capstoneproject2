#Business Understanding
## Project Overview

This project develops an end-to-end machine learning solution to predict whether a loan applicant is likely to default using the Home Credit Default Risk dataset. The workflow includes data preprocessing, exploratory data analysis, feature engineering, model building, and evaluation using classification algorithms such as Logistic Regression, Random Forest, and XGBoost. The final model is interpreted using SHAP for explainability and deployed through a Streamlit application with a Power BI dashboard, enabling financial institutions to make informed, data-driven lending decisions while reducing credit risk.


## Problem Statement

Financial institutions face significant financial losses when loans are approved for customers who are unable to repay them. Accurately identifying high-risk applicants before loan approval is essential for minimizing default risk while ensuring that eligible customers continue to receive credit. This project aims to develop a machine learning model that predicts the likelihood of loan default using customer demographic, financial, and credit history data, enabling more informed and data-driven lending decisions.


##Objective

Predict whether a loan applicant is likely to default on a loan so that financial institutions can make better lending decisions and reduce financial losses.
